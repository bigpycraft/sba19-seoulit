```python
import platform

import numpy as np
import pandas as pd

from konlpy.corpus import kobill
from konlpy.tag    import Twitter
from wordcloud     import WordCloud, STOPWORDS
from PIL import Image
from collections import Counter 

import nltk
import sys

files_ko = kobill.fileids()
with open("C://Users//OWNER//Anaconda_scr//twitter_data_2012-01-01_to_2012-12-30.csv") as my_file:
    text = my_file.read()
    text
    
    
#불용어처리 nltk
twitter = Twitter()

import nltk
nltk.download('punkt')

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 

#stop_words = 불필요한 단어 나열 (주의 간격 spacebar)
stop_words = "중국 이 등 것 생각 나선 후 때 그 사람 더 내 산 중 말 나 앞 놈 좀 뭐 일 불매운동 날 및 애플 분 봉 회 볼 하리 익금 위 로 거 팬 지금 진짜 이번 안 왜 때문 애 대해 미국 이제 수 요 년도 중국인"

#twitter.nouns(해당 text 넣기)
stop_words=stop_words.split(' ')
word_tokens = twitter.nouns(text)

#불용어를 제외한 명사만 모두 result에 추출
result = [] 
for w in word_tokens: 
    if w not in stop_words: 
        result.append(w) 

print(word_tokens) 
print(result) 
```
