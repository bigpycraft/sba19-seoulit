```python
land19 = land19.drop(land19.index[0:16])
land16 = land16.drop(land16.index[0:16])
land17 = land17.drop(land17.index[0:16])
```

```python
subset1 = land19[['□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.','Unnamed: 3','Unnamed: 7','Unnamed: 8']]
subset1.head(10)
```

```python
subset2 = land16[['□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.','Unnamed: 3','Unnamed: 7','Unnamed: 8']]
subset2.head(10)
```

```python
subset3 = land17[['□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.','Unnamed: 3','Unnamed: 7','Unnamed: 8']]
subset3.head(10)
```

```python
subset1 = subset1.rename(columns={'□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.' : '시군구',
                           'Unnamed: 7':'계약면적(㎡)',
                           'Unnamed: 3':'용도지역',
                           'Unnamed: 8':'거래금액(만원)'})
subset2 = subset2.rename(columns={'□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.' : '시군구',
                           'Unnamed: 7':'계약면적(㎡)',
                           'Unnamed: 3':'용도지역',
                           'Unnamed: 8':'거래금액(만원)'})
subset3 = subset3.rename(columns={'□ 본 서비스에서 제공하는 정보는 법적인 효력이 없으므로 참고용으로만 활용하시기 바랍니다.' : '시군구',
                           'Unnamed: 7':'계약면적(㎡)',
                           'Unnamed: 3':'용도지역',
                           'Unnamed: 8':'거래금액(만원)'})
```

```python
sc1 = len(subset1)
sc2 = len(subset2)
sc3 = len(subset3)
```

```python
for i in range(sc1):
    subset1['시군구'][i+16] = subset1['시군구'][i+16][0:9]
```

```python
for i in range(sc2):
    subset2['시군구'][i+16] = subset2['시군구'][i+16][0:9]
```

```python
for i in range(sc3):
    subset3['시군구'][i+16] = subset3['시군구'][i+16][0:9]
```

```python
subset1 = subset1[subset1['용도지역'].isin(['제1종일반주거지역', '제1종전용주거지역','제2종일반주거지역', '제3종일반주거지역'])]
subset2 = subset2[subset2['용도지역'].isin(['제1종일반주거지역', '제1종전용주거지역','제2종일반주거지역', '제3종일반주거지역'])]
subset3 = subset3[subset3['용도지역'].isin(['제1종일반주거지역', '제1종전용주거지역','제2종일반주거지역', '제3종일반주거지역'])]
```
