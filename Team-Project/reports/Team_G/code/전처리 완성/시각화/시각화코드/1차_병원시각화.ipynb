```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
sns.set_style('whitegrid')
sns.set_color_codes()
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="C:/Windows/Fonts/MALGUN.TTF").get_name()
rc('font', family=font_name)
plt.rcParams['figure.figsize'] = (15,10)
%matplotlib inline
```

```python
pop = pd.read_csv('../data/population_Final.csv',encoding='EUC-KR')
```

```python
hos = pd.read_excel('../data/의료_최종.xlsx',encoding='EUC-KR')
```

```python
pop
```

```python
hos = hos.rename(columns={'시/도':'행정구역'})
```

```python
hos = pd.merge(hos, pop, on='행정구역')
```

```python
hos = hos.drop(7,axis=0)
```

```python
hos
```

```python
# 전체 병원 수에서 차지하는 비율
hos['비율'] = hos['합계'] / hos['합계'].sum() * 100
```

```python
hos['평균인구대비'] = hos['합계'] / hos['인구수'].mean() * 100
```

```python
hos
```

```python
sns.barplot('행정구역','평균인구대비', data = hos, palette = 'Blues')
plt.title('평균인구대비 병원수')
plt.show()
```

```python
sns.barplot('행정구역','평균인구대비', data = hos, palette = 'RdBu')
plt.title('전체 병원대비 병원수')
plt.show()
```
