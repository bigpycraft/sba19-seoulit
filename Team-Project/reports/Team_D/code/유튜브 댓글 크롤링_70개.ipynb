```python
from selenium import webdriver

from urllib.parse import quote

import time

from selenium.webdriver.common.keys import Keys

from selenium.webdriver.support.ui import WebDriverWait

from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.common.by import By

import pandas as pd

import re

import sys
from bs4 import BeautifulSoup


def YoutubeGetDataFromUrl(browser,url,numofpage):

    browser.get(url)

    body=browser.find_element_by_tag_name('body')

    comment_list=[]

    content_list=[]

    num_page_down = 10
    while num_page_down:
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(1.5)
        num_page_down -= 1

    html = browser.page_source
    soup = BeautifulSoup(html, 'lxml')
    content_text = soup.select('#content-text')

    for i in list(range(len(content_text))):
        text = content_text[i].get_text()
        text = re.sub('\\n','',text)
        comment_list.append(text)
        comment_list+list(set(comment_list))

    return comment_list

        

browser=webdriver.Chrome('C:/selenium/chromedriver.exe')

url = 'https://www.youtube.com/watch?v=Y4R7seLmhHc'

comment=YoutubeGetDataFromUrl(browser,url,20)

print(comment)
```
