# 워드클라우드

```python
import pytagcloud
import pandas as pd
import pytagcloud as tagcloud
from IPython.display import Image 
```

```python
news_2015=pd.read_excel('2015_news_word_list.xlsx')
```

```python
news_2015.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
news_2015= news_2015.head(100)
```

```python
news_2015
```

```python
def df_tag(df):
    Result=[]
    for i in range(len(df)):
        word = (df['tag'][i],df['size'][i])
        Result.append(word)
        
    return Result
```

```python
result = df_tag(news_2015)
```

```python
tags = tagcloud.make_tags(result, maxsize=300)
import pprint
pprint.pprint(tags)
```

```python
def draw_cloud(tags, filename, fontname='Korean', size=(800,600)):
    pytagcloud.create_tag_image(tags, filename, fontname=fontname, size=size)
```

```python
img_wordcloud = 'wordcloud/news.png'
draw_cloud(tags, img_wordcloud, fontname='Korean', size=(1400, 1000))
Image(img_wordcloud)
```
