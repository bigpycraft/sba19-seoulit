```python
from bs4 import BeautifulSoup
from selenium import webdriver
from IPython.display import Image

import re
import pandas as pd
import time
```

```python
driver_path = "driver/chromedriver.exe"
driver = webdriver.Chrome(executable_path=driver_path)
```

```python
url_page = 'http://www.cgv.co.kr/theaters/?NaPm=ct%3Dk0d6vu9y%7Cci%3Dcheckout%7Ctr%3Dds%7Ctrx%3D%7Chk%3Df3431e03572314b7b15f7dee2194598954ad00f1'
time.sleep(2)
driver.get(url_page)
```

```python
area=[]
Addr=[]


for a1 in range(2,10):
    driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[2]/ul/li[{0}]/a'.format(a1)).click()
    driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[2]/ul/li[{0}]/div/ul/li[1]/a'.format(a1)).click()

    
    
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    soup.title

        # 광역시 내 지점명 전부 찾기
    ha= soup.find("li","on")
    name_all=ha.find_all('a')
    len(name_all)

    
    for a2 in range(2,len(name_all)-1):
        driver.find_element_by_xpath('//*[@id="contents"]/div[1]/div/div[2]/ul/li[{0}]/div/ul/li[{1}]/a'.format(a1,a2)).click()
        
        name = name_all[a2]    
        na = name.get_text()
        na
        
        
        html = driver.page_source
        soup = BeautifulSoup(html, "lxml")
        soup.title
    
    
        # 광역시 내 지점의 주소 찾기
        adr = soup.find('strong','title').get_text()
        ad = adr.split()
        ad1= ad[:3]
    
        
        area.append(na)
        Addr.append(ad1)
        
        
```

```python
area
```

```python
cgv2 = pd.DataFrame({'CGV매장':area,
                    '주소':Addr,})
```
