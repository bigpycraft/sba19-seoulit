```python
import json
import folium
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
```

```python
geo_path = './map.json'

geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
sum.rename(columns={'행정구역별(읍면동)':'지역'}, inplace=True)
sum.rename(columns={'2018':'인구수'}, inplace=True)
sum.head()
```

```python
guDat = pd.DataFrame({'gu':sum['지역'], 'counts':sum['인구수']})
guDat.head()
```

```python
guDat
```

```python
map = folium.Map(location=[35.5502, 126.982], zoom_start=11)

map.choropleth(geo_data=geo_str,
              data=guDat,
              columns=['gu', 'counts'],
              fill_color='YlGnBu', #PuRd, YlGnBu
              key_on='feature.id')
```
