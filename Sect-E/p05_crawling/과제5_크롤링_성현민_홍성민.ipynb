```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url="https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/"
resp = requests.get(url)
```

```python
soup=BeautifulSoup(resp.text,"lxml")
rank = soup.find_all('div','sammyRank')
bu_st = soup.find_all('div','sammyListing')
site=bu_st[0].find('a')['href']
type(site)
```

```python
# 가격, 주소, 전화번호, 홈페이지정보
Rank =[]
Burger =[]
Store =[]
Site = []
Link=[]
Price=[]
Address=[]
Phone=[]
rank_1=0
link=""

for i,j in enumerate(bu_st):
    tmp = bu_st[i].get_text()
    tmp = tmp.split('\r\n')
    tmp[1] = tmp[1].split('\n')[0]
    rank_1= int(i)+1
    site=bu_st[i].find('a')['href']
    if site.startswith('http://www.chicagomag.com'):
        Site.append(site)
    else:
        site='http://www.chicagomag.com'+site
        Site.append(site)
    
    Rank.append(rank_1)
    Burger.append(tmp[0])
    Store.append(tmp[1])
#     print(site)
    url_1=Site[i]
#     print(url_1)
    resp_1=requests.get(url_1)
    soup_1=BeautifulSoup(resp_1.text,"lxml")
    tmp_1 = soup_1.select("p.addy em")
#     print(tmp_1)
#     print(tmp_1[0].find('a')['href'])
# tmp_1[0].get_text()
    if 'http' in str(tmp_1[0]):
        link=tmp_1[0].find('a')['href']
    else:
        link="None"  
    Link.append(link)
    list=tmp_1[0].get_text().split(',')
    money=list[0].split(' ')
    if '-' in str(tmp_1[0]):
        phone=re.findall('[0-9]+-[0-9]+-[0-9]+', tmp_1[0].get_text())
    else:
        phone="None"
#     money_1 = 
    money_1=re.findall('[$0-9]+[.]?[0-9]+', tmp_1[0].get_text())
    Price.append(money_1[0])
    address=""
    for a,b in enumerate(money):
        if a>0:
            address+=" "+money[a]
    Address.append(address)
    Phone.append(phone[0])
#     print(Price)
#     print(Address)
#     print(Link)
    print(url_1)
    print(phone[0])
    print(money_1[0])
```

```python
data = {'Rank':Rank,'Burger':Burger,'Store':Store,"Site":Site,"Price":Price,"Phone":Phone,"Address":Address,"Link":Link}
data
```

```python
df = pd.DataFrame(data)
df
```
