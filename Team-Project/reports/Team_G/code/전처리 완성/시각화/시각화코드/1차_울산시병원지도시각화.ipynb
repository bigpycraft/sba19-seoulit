# 토지 실거래가 folium

```python
Hos = pd.read_csv('./data/토지매매가.csv', encoding='EUC-KR')
```

```python
Hos.head()
```

```python
Hos['시군구'] = Hos['시군구'].apply(lambda e: e.split()[1])
```

```python
Hos.head()
```

```python
GEO = Hos[['시군구','거래금액(만원)']]
```

```python
GEO = GEO.rename(columns={'거래금액(만원)' : '거래금액' })
```

```python
GEO.head()
```

```python
GEO['거래금액'].groupby(GEO['시군구']).mean()
```

```python
import folium
import json
latitude  = 35.5383773
longitude = 129.3113596
loc_tuple = (latitude, longitude)
loc_tuple
```

```python
geo_path = './data/skorea_municipalities_geo_simple1.json'
geo_str  = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
## 토지가격
map = folium.Map(location=[35.5383773, 129.3113596], tiles='Stamen Toner', zoom_start=11)

map.choropleth(geo_data=geo_str,
               data=Hos,
               columns=['시군구' ,'거래금액(만원)'],
               fill_color='YlGnBu', #PuRd, YlGnBu
               key_on='properties.name')
map
```

>- 토지매매가는 울주군, 중구, 동구가 가장 저렴
>- 시립병원 설립시 울주군 혹은 동구에 짓는 것이 가장 합리적이라 판단된다.
