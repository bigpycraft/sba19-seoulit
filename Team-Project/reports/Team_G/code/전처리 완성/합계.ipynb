```python
import pandas as pd
import glob
import re
```

```python
df_medi = pd.read_excel('./data/의료_통합.xlsx')
df_edu = pd.read_excel('./data/교육_통합.xlsx')
df_trans = pd.read_excel('./data/버스정류장_통합.xlsx')
```

```python
df_edu
```

```python
edu_sum = pd.DataFrame(df_edu.sum(axis=1), columns = ['합계'])
edu_sum
```

```python
edu = df_edu.join(edu_sum)
edu
```

```python
edu.sum()
```

```python
edu_sum2 = pd.DataFrame(edu.sum(), columns=['계'])
edu_sum2
```

```python
edu2  = edu.append(edu_sum2.T)
edu2
```

```python
edu2.loc['계', '지역']   = '전체'
edu2
```

```python
excel_file_name = './data/교육_최종.xlsx'

excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
edu2.to_excel(excel_total_file_writer, index=False, sheet_name='교육_통합')
excel_total_file_writer.save()

glob.glob(excel_file_name)
```

```python
df_trans
```

```python
trans_sum = pd.DataFrame(df_trans.sum(), columns=['계'])

trans  = df_trans.append(trans_sum.T)

trans.loc['계', '구분']   = '전체'
```

```python
trans
```

```python
excel_file_name = './data/버정_최종.xlsx'

excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
trans.to_excel(excel_total_file_writer, index=False, sheet_name='교통_통합')
excel_total_file_writer.save()

glob.glob(excel_file_name)
```
