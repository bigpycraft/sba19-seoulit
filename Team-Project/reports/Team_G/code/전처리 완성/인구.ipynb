```python
import pandas as pd
```

```python
pop = './data/201908_201908_주민등록인구및세대현황_월간.csv'
```

```python
pop
```

```python
df_pop = pd.read_csv(pop, engine="python")
```

```python
df_pop
```

```python
df_pop.loc[1, '행정구역'] = '서울'
df_pop.loc[2, '행정구역'] = '부산'
df_pop.loc[3, '행정구역'] = '대구'
df_pop.loc[4, '행정구역'] = '인천'
df_pop.loc[5, '행정구역'] = '광주'
df_pop.loc[6, '행정구역'] = '대전'
df_pop.loc[7, '행정구역'] = '울산'

df_pop
```

```python
new_pop = df_pop[['행정구역','2019년08월_총인구수']]
new_pop
```

```python
new_POP = new_pop[new_pop['행정구역'].
             isin(['서울','부산','대구','인천','광주','대전','울산'])]
new_POP
```

```python
import glob

excel_file_name = './data/인구_통합.xlsx'
excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
new_POP.to_excel(excel_total_file_writer, index=False, sheet_name='인구_통합')
excel_total_file_writer.save()

glob.glob(excel_file_name)
```

```python
df_pop_1 = pd.read_excel('./data/인구_통합.xlsx')
df_pop_1
```

```python
pop_sum = pd.DataFrame(df_pop_1.sum(), columns=['계'])
pop_sum
```

```python
pop2  = new_POP.append(pop_sum.T)
pop2
```

```python
pop2.loc['계', '행정구역']   = '전체'
pop2
```

```python
excel_file_name = './data/인구_최종.xlsx'

excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
pop2.to_excel(excel_total_file_writer, index=False, sheet_name='인구_통합')
excel_total_file_writer.save()

glob.glob(excel_file_name)
```
