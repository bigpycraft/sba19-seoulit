```python
dataset=[df["N"][0]];

for i in range(1,len(df)):
    dataset.append(df["N"][i])
```

```python
dataset[0][0]
```

```python
df1=pd.read_excel("keys_values_19_07.xlsx")
len(df1)
```

```python
keys=list(df1["keys"])
keys.append("일본")
keys.append("불매운동")
keys.append("불매")
```

```python
data=[]
for i in range(0,len(dataset)):
    data.append(0)
    
for i in range(0,len(dataset)):
    data[i]=[]
    for j in range(0,len(dataset[i])):
        if dataset[i][j] in keys:
            data[i].append(dataset[i][j])
```

```python
data[0]
```

```python
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)
fd=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
fd
```

```python
frequent_itemsets=apriori(fd,min_support=0.00009,use_colnames=True)  #지지도 0.00009
```

```python
frequent_itemsets
```

```python
from mlxtend.frequent_patterns import association_rules
```

### 항상도 lift가 1보다 작은 경우 우연적 기회보다 높은 확률

### 항상도 lift가 1인 경우 독립

### 항상도 lift가 1보다 큰 경우 우연적 기회보다 낮은 확률


```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)  #신뢰도 1
```
