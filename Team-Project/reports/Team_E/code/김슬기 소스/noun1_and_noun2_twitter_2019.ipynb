```python
import pandas as pd
```

```python
df=pd.read_excel("nouns_twitter_data_2019-07-01_to_2019-09-09.xlsx")
df.head()
```

```python
len(df)
```

```python
df["N"]='';

for i in range(0,len(df)):
    df["nouns"][i]=df["nouns"][i].replace("'","")
    df["nouns"][i]=df["nouns"][i].replace("[","")
    df["nouns"][i]=df["nouns"][i].replace("]","")
    df["nouns"][i]=df["nouns"][i].replace(" ","")
    df["N"][i]=df["nouns"][i].split(',')
```

```python
len(df["N"][0])
```

```python
noun1=[]

for i in range(0,len(df)):
    for j in range(0,len(df["N"][i])):
        for j in range(0,len(df["N"][i])):
            noun1.append(df["N"][i][j])
            
```

```python
noun2=[]

for i in range(0,len(df)):
    for j in range(0,len(df["N"][i])):
        a=df["N"][i][j]
        for j in range(0,len(df["N"][i])):
            noun2.append(a)
```

```python
noun2
```

```python
data=pd.DataFrame({"noun1":noun1,'noun2':noun2})
data.head(20)
```

```python
data.to_excel("noun1_and_noun2_twitter_data_2019-07-01_to_2019-09-09.xlsx")
```
