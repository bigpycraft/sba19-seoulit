```python
import pandas as pd
```

```python
medi = './data/의료.csv'
df_me = pd.read_csv(medi, engine="python")
```

```python
df_me
```

```python
new_me = df_me[df_me['시/도'].
             isin(['서울','부산','대구','인천','광주','대전','울산'])]
new_ME = new_me[['시/도','합계']]
```

```python
new_ME
```

```python
import glob

excel_file_name = './data/의료_통합.xlsx'
excel_total_file_writer = pd.ExcelWriter(excel_file_name, engine='xlsxwriter')
new_ME.to_excel(excel_total_file_writer, index=False, sheet_name='의료_통합')
excel_total_file_writer.save()

glob.glob(excel_file_name)
```
