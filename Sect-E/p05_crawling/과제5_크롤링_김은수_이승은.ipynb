```python
import glob
import time
from datetime import datetime

import os
from os.path import exists

import requests
from bs4 import BeautifulSoup

import re
import pandas as pd
```

```python
url='https://www.chicagomag.com/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago/'
```

```python
html1=requests.get(url).text
soup1=BeautifulSoup(html1,'lxml')

ranks=soup1.select('div.sammyRank')
menus=soup1.select('div.sammyListing b')
links=soup1.select('div.sammyListing a')
trains=soup1.select('div.sammyListing')

cafe=[]
for idx in range(len(trains)):
    nlis=trains[idx]
    nname=nlis.get_text().split('\n')[1]
    cafe.append(nname)

rank=[rank.get_text().strip() for rank in ranks]

menu=[menu.get_text().strip() for menu in menus]

link=[]
for idx in range(len(links)):
    if links[idx]['href'].startswith('h'):
        link.append(links[idx]['href'])
    else:
        link.append('http://www.chicagomag.com'+links[idx]['href'])
```

```python
data_new={'Rank':rank,'Cafe':cafe,'Menu':menu,'Link':link}
df=pd.DataFrame(data_new)
df.head(10)
```

```python
Cost=[]
Address=[]
PNUM=[]
HP=[]
for lk in link:
    url=lk
    html2=requests.get(url).text
    soup2=BeautifulSoup(html2,'lxml')
    
    
    adds=soup2.select('p.addy em')
    v=adds[0].get_text()

    cost=v.split('. ')[0]

    pnpat=re.compile('\d+[-]\d+[-]\d+')
    hpat1=re.compile('\w+[.]com')
    hpat2=re.compile('\w+[.]net')

    hp1=hpat1.findall(v)
    hp2=hpat2.findall(v)
    
    street=v.split('.,')[0].split('. ')[1]
    #street
    state=v.split('.,')[0].split('.')[-1]
    #state
    
    
    
    if pnpat.findall(v):
        pn=pnpat.findall(v)
        pn=pn[0]
        add=street+'.'+state
    else:
        pn=None
        add='Multiple Location'
    
    if hp1:
        hp=hp1[0]
    elif hp2:
        hp=hp2[0]
    else:
        hp=' '

    

    
    
    
    
    
    
#    adds=soup2.select('p.addy em')
#    v=adds[0].get_text()
#    cost=v.split('. ')[0] ####
#    add=v.split(',')[0].split('. ')[1]
#    pns=v.split(',')[-2] ###
#    hp=v.split(', ')[-1]
    Cost.append(cost)
    Address.append(add)
    PNUM.append(pn)
    HP.append(hp)
```

```python
df['Cost']=Cost
df['Address']=Address
df['PNUM']=PNUM
df['HP']=HP
df
```
