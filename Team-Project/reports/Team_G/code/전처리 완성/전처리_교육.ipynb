```python
high = './data/high.csv'
mid = './data/중학교.csv'
ele = './data/초등학교.csv'

df_h = pd.read_csv(high, engine="python")
df_m = pd.read_csv(mid, engine="python")
df_e = pd.read_csv(ele, engine="python")
```

```python
df_e.loc[1, '행정구역별'] = '서울'
df_e.loc[2, '행정구역별'] = '부산'
df_e.loc[3, '행정구역별'] = '대구'
df_e.loc[4, '행정구역별'] = '인천'
df_e.loc[5, '행정구역별'] = '광주'
df_e.loc[6, '행정구역별'] = '대전'
df_e.loc[7, '행정구역별'] = '울산'

df_e
```

```python
new_e = df_e[df_e['행정구역별'].
             isin(['서울','부산','대구','인천','광주','대전','울산'])]
new_E = new_e[['행정구역별','2018']]
```

```python
new_E
```

```python
df_E = pd.DataFrame(new_E)
df_E = df_E.sort_values(['행정구역별'], ascending=[True])
df_E
```

```python
ele_list = list(df_E['2018'])
ele_list
```

```python
data = {'지역':reg,
       '고등학교':high_list,
       '중학교':mid_list,
       '초등학교':ele_list}
edu = pd.DataFrame(data)
```
